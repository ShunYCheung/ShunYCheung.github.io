```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib as mpl

mpl.rcParams['font.family'] = 'serif'   # or 'sans-serif', 'monospace'
mpl.rcParams['font.serif'] = ['Times New Roman']   # choose your preferred font
mpl.rcParams['font.size'] = 16          # default font size
```

```python
x1 = np.linspace(-5, -1, 20)
x2 = np.linspace(-1, 1, 10)
x3 = np.linspace(1, 5, 20)

y1d = np.zeros(70)
```

```python
plt.scatter(x1, np.zeros(len(x1)), color='blue')
plt.scatter(x2, np.zeros(len(x2)), color='red')
plt.scatter(x3, np.zeros(len(x3)), color='blue')
plt.yticks([])
```

```python
y1 = x1**2
y2 = x2**2
y3 = x3**2
plt.scatter(x1, y1, color='blue')
plt.scatter(x2, y2, color='red')
plt.scatter(x3, y3, color='blue')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].scatter(x1, np.zeros(len(x1)), color='blue')
ax[0].scatter(x2, np.zeros(len(x2)), color='red')
ax[0].scatter(x3, np.zeros(len(x3)), color='blue')
ax[0].set_yticks([])

ax[1].scatter(x1, y1, color='blue')
ax[1].scatter(x2, y2, color='red')
ax[1].scatter(x3, y3, color='blue')

plt.axhline(1)

plt.savefig('kerneltrick.png')
```
