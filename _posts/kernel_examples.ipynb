```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

plt.rcParams.update({
    "font.size": 18,  })
```

```python
# Hyperparameters
sigma = 1.0
ell = 0.2

x = np.linspace(0, 1, 200)
X1, X2 = np.meshgrid(x, x)
K = sigma**2 * np.exp(-0.5 * ((X1 - X2)/ell)**2)

# Sample from GP
y = np.random.multivariate_normal(mean=np.zeros_like(x), cov=K, size=1)

plt.figure(figsize=(8,5))
plt.plot(x, y.T, linewidth=1)
plt.title("Sample Functions from RBF Kernel GP")
plt.xlabel("x")
plt.ylabel("k(x)")
plt.show()
```

```python
fig= plt.figure(figsize=(8,5))

def animate(i):
    
    plt.cla()
    # Sample from GP
    y = np.random.multivariate_normal(mean=np.zeros_like(x), cov=K, size=1)
    plt.plot(x, y.T, linewidth=1)
    plt.title("RBF Kernel")
    plt.xlabel("x")
    plt.ylabel("k(x)")
    plt.ylim(-3, 3)
    plt.tight_layout()
ani = animation.FuncAnimation(fig, animate, repeat=True, frames=10, interval=1000)
writer = animation.PillowWriter(fps=1.5,
                                metadata=dict(artist='Me'),
                                bitrate=1800)
ani.save('rbf_kernel_example.gif', writer=writer)
```

```python
# Kernel hyperparameters
sigma = 1.0
ell = 1

# Distances
r = np.linspace(0, 1, 200)
x = np.linspace(-3, 3, 200)
# Matern 3/2 kernel

X1, X2 = np.meshgrid(x, x)
K = sigma**2 * (1 + np.sqrt(3) * np.abs(X1-X2) / ell) * np.exp(-np.sqrt(3) * np.abs(X1-X2) / ell)

fig= plt.figure(figsize=(8,5))

def animate(i):
    
    plt.cla()
    # Sample from GP
    y = np.random.multivariate_normal(mean=np.zeros_like(x), cov=K, size=1)
    plt.plot(x, y.T, linewidth=1)
    plt.title("Matern 3/2 Kernel")
    plt.xlabel("x")
    plt.ylabel("k(x)")
    plt.ylim(-3, 3)
    plt.tight_layout()
ani = animation.FuncAnimation(fig, animate, repeat=True, frames=10, interval=1000)
writer = animation.PillowWriter(fps=1.5,
                                metadata=dict(artist='Me'),
                                bitrate=1800)
ani.save('matern_3_2_kernel_example.gif', writer=writer)
```

```python
# Kernel hyperparameters
sigma = 1.0
ell = 1
p=1

# Distances
r = np.linspace(0, 1, 200)
x = np.linspace(-3, 3, 200)
# Matern 3/2 kernel

X1, X2 = np.meshgrid(x, x)
K = sigma**2 * np.exp(-2*np.sin(np.pi*np.abs((X1-X2))/p)**2/ell)

fig= plt.figure(figsize=(8,5))

def animate(i):
    
    plt.cla()
    # Sample from GP
    y = np.random.multivariate_normal(mean=np.zeros_like(x), cov=K, size=1)
    plt.plot(x, y.T, linewidth=1)
    plt.title("Periodic Kernel")
    plt.xlabel("x")
    plt.ylabel("k(x)")
    plt.ylim(-3, 3)
    plt.tight_layout()
ani = animation.FuncAnimation(fig, animate, repeat=True, frames=10, interval=1000)
writer = animation.PillowWriter(fps=1.5,
                                metadata=dict(artist='Me'),
                                bitrate=1800)
ani.save('periodic_kernel_example.gif', writer=writer)
```
