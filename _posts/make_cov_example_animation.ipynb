```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots(ncols = 2, figsize=(14, 6))
def animate(i):
    plt.cla()
    # Grid of points
    x = np.linspace(-3, 3, 300)
    y = np.linspace(-3, 3, 300)
    X, Y = np.meshgrid(x, y)

    # Gaussian parameters
    mu_x, mu_y = 0, 0
    sigma_x, sigma_y = 1.0, 1.5

    # 2D Gaussian function
    Z = np.exp(
        -0.5 * (
            ((X - mu_x)/sigma_x)**2 +
            ((Y - mu_y)/sigma_y)**2
        )
    )
    ax[0].contourf(X, Y, Z, levels=3)

    x0 = np.random.normal(0, 1)
    y0 = np.random.normal(0, 1)
    ax[0].contour(X, Y, Z, colors='k', linewidths=0.5, levels=3)
    ax[0].scatter(x0, y0,marker='o', linestyle='None', color='black')
    ax[0].set_xlabel(r'$y_1$')
    ax[0].set_ylabel(r'$y_2$')
    ax[1].plot([1, 2], [x0, y0], marker='o')
    ax[1].set_xlim(0, 3)
    ax[1].set_ylim(-3, 3)
    ax[1].set_xlabel(r'$x$')
    ax[1].set_ylabel(r'$y$')
    return (ax,)

ani = animation.FuncAnimation(fig, animate, repeat=True, frames=10, interval=1000)
writer = animation.PillowWriter(fps=1.5,
                                metadata=dict(artist='Me'),
                                bitrate=1800)
ani.save('correlation_2_points.gif', writer=writer)
```
